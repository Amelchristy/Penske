```python
import pandas as pd

# Define only numeric columns that should be averaged
numeric_columns_to_average = [
    'N_JOBS', 'N_WEEKEND_DAYS', 'N_WEEKS', 'TOTAL_HOURS', 'WORKED_HOURS', 'REG_HRS', 'REG_AMT',
    'OT_HRS', 'OT_AMT', 'VACATION_HRS', 'SICK_HRS', 'NON_WORKED_HRS', 'BONUS', 'INCENTIVE',
    'SHIFT_DIFF', 'LEAD', 'CHECK_GROSS', 'AVG_TOTAL_HOURS', 'AVG_WORKED_HOURS', 'AVG_REG_HRS',
    'AVG_REG_AMT', 'AVG_OT_HRS', 'AVG_OT_AMT', 'AVG_SICK_HRS', 'AVG_CHECK_GROSS', 'AVG_N_WEEKEND_DAYS',
    'TIP', 'AVG_TOTAL_HOURS_PCT_CHG', 'AVG_WORKED_HOURS_PCT_CHG', 'AVG_OT_HRS_PCT_CHG',
    'AVG_REG_HRS_PCT_CHG', 'AVG_REG_AMT_PCT_CHG', 'AVG_OT_HRS_PCT_AMT', 'AVG_SICK_HRS_PCT_CHG',
    'AVG_CHECK_GROSS_PCT_CHG', 'START_STD', 'END_STD', 'START_HOUR_AVG', 'END_HOUR_AVG'
]

# Group data by ANON_SSO and aggregate
aggregated_data = penske_df.groupby('ANON_SSO').agg({
    **{col: 'mean' for col in numeric_columns_to_average},  # Mean for numeric columns
    'STATUS': lambda x: 'T' if 'T' in x.values else 'E',  # Determine final status
    'YEAR_MONTH': 'max'  # Presumably you might want the latest date
})

# Add aggregation for dummy variables dynamically from the DataFrame
area_columns = [col for col in penske_df.columns if 'AREA_NAME' in col]
for col in area_columns:
    aggregated_data[col] = penske_df.groupby('ANON_SSO')[col].max()

# Reset index to make ANON_SSO a column again
aggregated_data.reset_index(inplace=True)

# Display the first few rows of the aggregated data to confirm changes
print(aggregated_data.head())

```

```python
#Building D matrice for XGBoost
from sklearn.model_selection import train_test_split

# Extract feature and target arrays
X, y = penske_df.drop('STATUS', axis=1), penske_df[['STATUS']]
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.4, random_state=1)
```

```python


# Initialize the models
algos = [
    GradientBoostingClassifier(random_state=1),
    XGBClassifier(random_state=1),
    MLPClassifier(activation='relu', hidden_layer_sizes=(50,18,1), max_iter=500, random_state=1)
]

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

results = []
for algo in algos:
    algo.fit(X_train, y_train)
    y_pred = algo.predict(X_test)
    metrics = {
        'Algorithm': algo.__class__.__name__,
        'Balanced Accuracy': balanced_accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'Accuracy': accuracy_score(y_test, y_pred)
    }
    results.append(metrics)

# Convert results into a DataFrame for nicer display
results_df = pd.DataFrame(results)
print(results_df)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import GridSearchCV
# Feature selection with RFECV to identify the most important features
xgb_estimator = XGBClassifier(objective='binary:logistic', random_state=1)
rfecv = RFECV(estimator=xgb_estimator, step=1, cv=3, scoring='balanced_accuracy')
rfecv.fit(X_train, y_train)
X_train_selected = X_train.loc[:, rfecv.support_]
X_test_selected = X_test.loc[:, rfecv.support_]

# Train an initial model with the selected features
xgb_initial = XGBClassifier(random_state=1)
xgb_initial.fit(X_train_selected, y_train)

# Predict and evaluate the initial model
y_pred_initial = xgb_initial.predict(X_test_selected)
initial_bal_acc = balanced_accuracy_score(y_test, y_pred_initial)
initial_acc = accuracy_score(y_test, y_pred_initial)
print(f"Initial Balanced Accuracy for RFE: {initial_bal_acc}")
print(f"Initial Accuracy for RFE: {initial_acc}")

# Define the parameter grid for GridSearchCV, including the 'gamma' parameter
param_grid = {
    'max_depth': [3, 5, 7, 9],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'gamma': [0, 0.01, 0.1, 0.2, 0.3]  
}

# Execute GridSearchCV on the selected features to find the best model parameters
grid_search = GridSearchCV(estimator=XGBClassifier(objective='binary:logistic', random_state=1),
                           param_grid=param_grid, scoring='balanced_accuracy', cv=3, verbose=1)
grid_search.fit(X_train_selected, y_train)

# Predict on the test set using the best model found by GridSearchCV
y_pred_optimized = grid_search.best_estimator_.predict(X_test_selected)

# Calculate both normal and balanced accuracy for the optimized model
optimized_acc = accuracy_score(y_test, y_pred_optimized)
optimized_bal_acc = balanced_accuracy_score(y_test, y_pred_optimized)
print(f"Optimized Accuracy: {optimized_acc}")
print(f"Optimized Balanced Accuracy: {optimized_bal_acc}")

# Print the best parameters and the best score obtained from GridSearchCV
print("Best parameters from GridSearchCV:", grid_search.best_params_)
print("Best balanced accuracy from CV:", grid_search.best_score_)
```

```python
import shap
from xgboost import XGBClassifier

# Initialize and fit the model with the best parameters found by GridSearchCV
xgb_model = grid_search.best_estimator_
xgb_model.fit(X_train_selected, y_train)  # Ensure the model is retrained for completeness

# Compute SHAP values
explainer = shap.TreeExplainer(xgb_model)
shap_values = explainer.shap_values(X_test_selected)  # Use X_test_selected to align with the trained model

# Visualize the feature importance using SHAP summary plot
shap.summary_plot(shap_values, X_test_selected, plot_type="dot")

# Visualize the feature importance using SHAP summary bar plot
shap.summary_plot(shap_values, X_test_selected, plot_type="bar")


```

```python
shap.plots.waterfall(shap_values[0])

```

```python
# Calculate the mean absolute SHAP values for each feature
shap_sum = np.abs(shap_values).mean(axis=0)
feature_importance = pd.DataFrame(list(zip(X_test.columns, shap_sum)), columns=['col_name', 'shap_importance'])
feature_importance.sort_values('shap_importance', ascending=False, inplace=True)

# Display the most important features
print(feature_importance)
```
